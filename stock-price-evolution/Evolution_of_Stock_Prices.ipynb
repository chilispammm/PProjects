```python

import numpy as np # Working with arrays
import matplotlib.pyplot as plt # For Data Viz

# Parameters
S0 = 100        # Initial stock price
mu = 0.05       # Drift
sigma = 0.2     # Volatility
T = 1.0         # Time horizon
dt = 0.01       # Time increment
N = int(T / dt) # Number of steps
t = np.linspace(0, T, N+1)

# Stock price array
S = np.zeros(N+1)
S[0] = S0

# Simulation of Brownian motion and stock prices
for i in range(1, N+1):
    dW = np.random.normal(0, np.sqrt(dt))
    S[i] = S[i-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * dW)

# A plot of the stock price path
plt.figure(figsize=(10,6))
plt.plot(t, S, label=f'Stock Price S(t)')
plt.title('Simulated Stock Price SDE Solution')
plt.xlabel('Time (t)')
plt.ylabel('Stock Price (S)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
S0 = 100  # Initial stock price
mu = 0.05  # Drift
sigma = 0.2  # Volatility
T = 1.0  # Time horizon
dt = 0.01  # Time increment
N = int(T / dt)  # Number of steps
t = np.linspace(0, T, N+1)

# Stock price array
S = np.zeros(N+1)
S[0] = S0

# Simulation of Brownian motion and stock prices
for i in range(1, N+1):
    dW = np.random.normal(0, np.sqrt(dt))
    S[i] = S[i-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * dW)

# A plot of the stock price path
plt.figure(figsize=(10,6))
plt.plot(t, S, label=f'Stock Price S(t)')
plt.title('Simulated Stock Price SDE Solution')
plt.xlabel('Time (t)')
plt.ylabel('Stock Price (S)')
plt.grid(True)
plt.legend()
plt.show()
```

## Key Insights

This notebook simulates a stock price path using a Stochastic Differential Equation (SDE), specifically the Geometric Brownian Motion model.

*   **Model Used:** Geometric Brownian Motion (GBM), a common model for financial asset prices.
*   **Parameters:** The simulation uses initial stock price ($S_0$), drift ($\mu$), volatility ($\sigma$), time horizon ($T$), and time increment ($dt$).
*   **Method:** The simulation is performed using an Euler-Maruyama scheme to discretize the SDE.
*   **Result:** The plot shows a single simulated path of the stock price over time, illustrating how the price evolves based on the defined parameters and random fluctuations.

## Key Insights

This notebook demonstrates a numerical simulation of a stochastic process frequently employed in financial modeling: Geometric Brownian Motion (GBM). The simulation utilizes the Euler-Maruyama method to approximate the solution of the associated Stochastic Differential Equation (SDE).

*   **Model:** The underlying model is the Geometric Brownian Motion, characterized by a constant drift ($\mu$) and volatility ($\sigma$). The SDE is given by $dS_t = \mu S_t dt + \sigma S_t dW_t$, where $dW_t$ represents a Wiener process increment.
*   **Discretization:** The Euler-Maruyama scheme approximates the continuous-time process by stepping through discrete time intervals ($\Delta t$), using the formula $S_{t+\Delta t} = S_t + \mu S_t \Delta t + \sigma S_t \sqrt{\Delta t} Z_t$, where $Z_t$ is a standard normal random variable.
*   **Parameters:** The simulation is driven by parameters that influence the trajectory's expected return and volatility: initial price ($S_0$), drift ($\mu$), volatility ($\sigma$), and the time step ($dt$).
*   **Output:** The generated plot represents one realization of this stochastic process. The path's characteristics, such as its overall trend and fluctuations, are a direct consequence of the chosen parameters and the inherent randomness of the Wiener process. The visual output provides an illustration of how the interplay between deterministic drift and stochastic volatility shapes the simulated asset price over time.
